Import all libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly
%config InlineBackend.figure_formats = ['svg']
```

## Import Data

```python
# generate test data

exp_A = np.concatenate((np.random.exponential(scale=1, size=100000), np.random.normal(loc=5,scale=0.2, size=100000)))
exp_B = np.concatenate((np.random.exponential(scale=0.5, size=50000), np.random.uniform(low=0,high=15, size=20000)))
exp_C = np.concatenate((np.random.exponential(scale=2, size=150000), np.random.normal(loc=10,scale=2, size=150000)))
```

## Plot

```python
# plot parameters - adapt them as you like

bins=500
ran=(0,20)
ylim = (0.5, 10000) 
xlim = (0,20)
yscale = 'log'  # try 'log' or 'linear'
```

```python
# exposure in kg days

exposure_A = 1
exposure_B = 1
exposure_C = 1
```

```python
# make histograms

hist_A, bins_A = np.histogram(exp_A, bins, range=ran)
hist_B, bins_B = np.histogram(exp_B, bins, range=ran)
hist_C, bins_C = np.histogram(exp_C, bins, range=ran)
```

```python
# plot - comment the lines of experiments to not show them

plt.hist(bins_A[:-1], bins_A, weights=hist_A/exposure_A, histtype='step', label='Experiment A')
plt.hist(bins_B[:-1], bins_B, weights=hist_B/exposure_B, histtype='step', label='Experiment B')
plt.hist(bins_C[:-1], bins_C, weights=hist_C/exposure_C, histtype='step', label='Experiment C')
plt.xlabel('Energy (keV)')
plt.ylabel(f'Counts / ({1000*(ran[1] - ran[0])/bins:.1f} eV kg days)')
plt.legend()
plt.xlim(xlim)
plt.ylim(ylim)
plt.yscale(yscale)
plt.show()
```
