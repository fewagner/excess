```python
# imports

import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data

data = np.loadtxt('../../original_data/Edelweiss/SpectrumRed30_Corrected-scalar-keVee.txt')
```

```python
# conversion

converted_data = np.empty(data.shape, dtype=float)

converted_data[:, 0] = data[:, 0] - data[:, 2]/2
converted_data[:, 1] = data[:, 0] + data[:, 2]/2
converted_data[:-1, 1] = converted_data[1:, 0]
converted_data[:, 2] = data[:, 1]
converted_data
```

```python
# write data to new file

np.savetxt('RED30_keVee.txt', converted_data)
```
